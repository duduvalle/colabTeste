```python
import wisardpkg as wp
import csv
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.utils import shuffle
import time
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
x_treino = []
y_treino = []

x_teste = []
y_teste = []

with open('dota2Train.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
      x_auxTreino = [0] * 226 
      for i in range(0,117):
        if i == 0:
          y_treino.append(row[i])
        elif i >= 4:
          if row[i] == "1":
            x_auxTreino[i-4] = 1
          elif row[i] == "-1":
            x_auxTreino[i+109] = 1
      x_treino.append(x_auxTreino)

with open('dota2Test.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
      x_auxTeste = [0] * 226 
      for i in range(0,117):
        if i == 0:
          y_teste.append(row[i])
        elif i >= 4:
          if row[i] == "1":
            x_auxTeste[i-4] = 1
          elif row[i] == "-1":
            x_auxTeste[i+109] = 1
      x_teste.append(x_auxTeste)
```

```python
print(len(x_treino))
print(len(x_teste))
print(x_treino[0])
print(y_treino[0])
```

```python
def ksplit(X, y, k=10):
    n = len(y)
    p = math.floor(n/k)

    Xs, ys = [X[p*i: p*(i+1)] for i in range(k)], [y[p*i: p*(i+1)] for i in range(k)] 
    return Xs, ys

x_treinoRandomizado, y_treinoRandomizado = shuffle(x_treino, y_treino)

print(len(x_treinoRandomizado))
print(x_treinoRandomizado[0])
print(y_treinoRandomizado[0])

x_split, y_split = ksplit(x_treinoRandomizado, y_treinoRandomizado)
```

```python
def accuracy(y_pred, y_target):
  return np.where(np.array(y_pred) == np.array(y_target), 1, 0).sum()/len(y_target)
```

```python
f = open("drive/MyDrive/1RSP-TEEDC/resultadosBleach.txt", "w")

for i in range(2,65):
  mediaAcuracia = 0
  mediaTempo = 0
  melhorTempo = 0
  melhorAcuracia = 0
  for j in range(10):
    melhorAcuraciaCros = 0
    model = wp.Wisard(i, ignoreZero = False, bleachingActivated=True)
    start_tempo = time.time()

    for k in range(10):
      model.train(x_split[k], y_split[k])
    
    for k in range(10):
      model.leaveMoreOut(x_split[k], y_split[k])
      resultadoAux = accuracy(model.classify(x_teste), y_teste)
      if resultadoAux > melhorAcuraciaCros:
        melhorAcuraciaCros = resultadoAux
      model.train(x_split[k], y_split[k])   

    tempo = time.time() - start_tempo
    resultado = melhorAcuraciaCros
    mediaAcuracia += resultado
    mediaTempo += tempo
    print(tempo)
    if resultado > melhorAcuracia:
      melhorTempo = tempo
      melhorAcuracia = resultado
  mediaTempo = mediaTempo/10
  mediaAcuracia = mediaAcuracia/10
  f.write(f"{mediaAcuracia}, {mediaTempo},\n")
  f.flush()
  print(f"{i} Melhor acuracia: {melhorAcuracia} -- Melhor tempo: {melhorTempo}")
  print(f"{mediaAcuracia} , {mediaTempo}")
```

```python
melhorTempo = 0
melhorAcuracia = 0
for i in range(1000):
    model = wp.Wisard(2, ignoreZero = True, bleachingActivated=True)
    start_tempo = time.time()
    model.train(x_treino, y_treino)  
    tempo = time.time() - start_tempo
    resultado = accuracy(model.classify(x_teste), y_teste)
    if resultado > melhorAcuracia:
      melhorTempo = tempo
      melhorAcuracia = resultado
      print(f"{i} acuracia: {melhorAcuracia} com tempo {melhorTempo}")
print(f"Melhor Acuracia: {melhorAcuracia} com tempo {melhorTempo}")
```
